```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
import requests
import json

def obter_comentarios(video_id, api_key):
    # Obtendo detalhes do vídeo
    video_info_url = f"https://www.googleapis.com/youtube/v3/videos?id={video_id}&part=snippet&key={api_key}"
    video_info_response = requests.get(video_info_url)
    video_info_data = json.loads(video_info_response.text)

    if 'items' in video_info_data:
        video_info = video_info_data['items'][0]['snippet']
        nome_do_video = video_info['title']
        data_de_publicacao = video_info['publishedAt']
        print(f"Nome do vídeo: {nome_do_video}")
        print(f"Data de publicação: {data_de_publicacao}")
        print("--------------------------")

        # Obtendo comentários do vídeo
        nextPageToken = None

        while True:
            url = f"https://www.googleapis.com/youtube/v3/commentThreads?videoId={video_id}&part=snippet&key={api_key}&maxResults=100"

            if nextPageToken:
                url += f"&pageToken={nextPageToken}"

            response = requests.get(url)
            data = json.loads(response.text)

            if 'items' in data:
                for comentario in data['items']:
                    texto_comentario = comentario['snippet']['topLevelComment']['snippet']['textDisplay']
                    data_comentario = comentario['snippet']['topLevelComment']['snippet']['publishedAt']

                    print(f"Comentário: {texto_comentario}")
                    print(f"Data do comentário: {data_comentario}")
                    print("--------------------------")

                if 'nextPageToken' in data:
                    nextPageToken = data['nextPageToken']
                else:
                    break
            else:
                print("Nenhum comentário encontrado para este vídeo.")
                break
    else:
        print("Vídeo não encontrado.")

# Substitua 'SEU_VIDEO_ID' pela ID do vídeo que você quer obter informações
video_id = '_UvZaHh1iZ4'

# Substitua 'SUA_CHAVE_DE_API' pela sua chave de API do YouTube
api_key = 'AIzaSyAnvjoHWW3Cu3jG5Z7bUFKuyiq7kBp0EXo'

# Chamando a função para obter informações do vídeo e seus comentários
obter_comentarios(video_id, api_key)
```
